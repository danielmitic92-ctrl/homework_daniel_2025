```python
import util
file_path='/Users/danielmitic/Desktop/homework_daniel_2025/aoc25/advent_of_code_3.txt'
input_file=util.read_file_into_lines(file_path)
```

```python
def find_battery(battery):
    #bank as list
    bank=list(battery)

    #find largest number
    first_number=max(bank[:-1]) #exlude last number bc it cant be the first digit

    #index of largest number
    index_first_number=bank.index(first_number)


    #determine second number which is the largest number after the max
    new_bank=bank[index_first_number+1:]
    second_number=max(new_bank)

    #combine first and second number
    combined=first_number+second_number
    return int(combined)   
```

```python
all_batteries=0
for battery in input_file:
    single_battery=find_battery(battery)
    all_batteries+=single_battery
```
