## 1. define functions

```python
def read_file_into_lines(file_path):
    """"
    read file and remove the character at the end
    """
    lines=[]
    with open (file_path,"r") as infile:
      for line in infile.readlines():
          cleaned=line.rstrip()
          lines.append(cleaned)
    return lines
```

```python
def compare_first_and_second_part(word):
    word=str(word)
    if len(word)%2!=0: #if uneven return 0
        return 0
    if len(word)%2==0: # if even compare the two halves
        first_half=word[0:int(len(word)/2)]
        second_half=word[int(len(word)/2):int(len(word))]
    if first_half==second_half:
        return word
    else:
        return 0 #if the two halfes arenot equal return 0
```

```python
def add_invalid_ID_in_range(lower, higher):
    "return the sum of all invalid IDs within a given range"
    range_sum=0
    for ID in range(lower,higher+1):
        invalid_ID=compare_first_and_second_part(ID)
        range_sum+=int(invalid_ID)
    return(range_sum)
```

## 2. load input file and do some preprocessing

```python
file_path='/Users/danielmitic/Desktop/phd/phd_29_07_2025/courses/Practical introduction to programming for biologists/advent_of_code_2.txt'
IDs=read_file_into_lines(file_path)
all_IDs=IDs[0].split(",")
```

## 3. applying functions to solve the exercise

```python
sum_off_all_wrong_IDs=0 #counter for the final sum
for ID in all_IDs: #looping over all IDs
    single_range=ID.split("-") #getting input for function
    lower=int(single_range[0]) #getting input for function
    higher=int(single_range[1]) #getting input for function
    
    sum_of_single_range=add_invalid_ID_in_range(lower,higher) ##calulating sums for individual ranges
    
    sum_off_all_wrong_IDs=sum_off_all_wrong_IDs+sum_of_single_range ##adding sums of the individual ranges
print(sum_off_all_wrong_IDs)
```

## 4. putting it all together in a single function
