```python
import util
file_path='/Users/danielmitic/Desktop/phd/phd_29_07_2025/courses/Practical introduction to programming for biologists/advent_of_code_1.txt'
rotations=util.read_file_into_lines(file_path)
```

```python
rotations=["L68","L30","R48","L5","R60","L55","L1","L99","R14","L23"]
```

```python
counter=0
current_postion=50
for rotation in rotations:
    if rotation.startswith("L"): 
        amount=int(rotation[1:])
        current_postion=current_postion-amount
        while current_postion < 0:
            current_postion=100+current_postion
        
    if rotation.startswith("R"):
        amount=int(rotation[1:])
        current_postion=current_postion+amount
        while current_postion>=100:
            current_postion=current_postion-100
    #print(current_postion)
            
    if current_postion==0:
        counter+=1
```
