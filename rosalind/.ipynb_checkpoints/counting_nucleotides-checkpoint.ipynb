```python
## solving the exercise unsing a loop
dna="GCATGGAACTA"

counter_A=0
counter_C=0
counter_T=0
counter_G=0

for base in dna:
    if base=="A":
        counter_A=counter_A+1
    if base=="C":
        counter_C=counter_C+1
    if base=="T":
        counter_T=counter_T+1
    if base=="G":
        counter_G=counter_G+1
print("A:", counter_A, "C:", counter_C, "G:", counter_G, "T:", counter_G)
```
