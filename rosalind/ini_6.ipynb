```python
#make dictionary
word_counts={}
for word in s.split(' '):
    word_counts[word]=0
   
```

```python
for key, value in word_counts.items():
    counter=0
    for word in s.split(' '):
        if word ==key:
            counter=counter+1
    #print(key)
    #print(counter)
    word_counts[key]=counter
```
