# FIB- Rabbits and Recurrence Relations 
https://rosalind.info/problems/fib/

```python
n=28
k=2
```

```python
start_sequence=[1,1]
```

```python
for month in range(n):
    if month>=2: #making sure not to have a negativ index
        reproductive_rabbits=start_sequence[month-2]
        baby_rabbits=k*reproductive_rabbits
        existing_rabbits=start_sequence[month-1]
        start_sequence.append(baby_rabbits + existing_rabbits)
print(start_sequence[-1])
```
