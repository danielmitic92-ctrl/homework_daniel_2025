# Computing GC content

```python
#1 import input file
file_path="/Users/danielmitic/Downloads/rosalind_gc.txt"
import util
fasta=util.read_file_into_lines(file_path)

```

```python
fasta
```

```python
#2 preprocessing of input file
sequence="".join(fasta) #join list to one string
sequence=sequence.split(">")[1:]#split along ID, exluding first line. 
```

```python
#calculate gc content for each sequence and save in dict
gc_contents={}
for line in sequence:

    #separte Id and sequence
    ID=line[0:13]
    bases=line[13:]
    
    #calculate GC content:
    GC_percentage=(bases.count("G") + bases.count("C"))/len(bases)*100

    #save results in dicts
    gc_contents[ID]=GC_percentage
```

```python
# determine ID with highest gc content
highest_GC=max(gc_contents.values())
for key,value in gc_contents.items():
    if value==highest_GC:
        print(key)
        print(value)
```
